```python
import numpy as np
import matplotlib.pyplot as plt
from  matplotlib import patches
from matplotlib.figure import Figure
from matplotlib import rcParams
from scipy.signal import lfilter
from scipy.fftpack import fft,fftfreq
```

```python
#(a)
fs = 10**3
A = [0.5,5,50]
N = 1024
t = np.arange(N)python
omega = N/fs
xn_seq = np.array([np.exp(-a*t*(10**-3)) for a in A])

for xn in xn_seq:
    plt.plot(xn)
plt.xlabel('Time index n')
plt.ylabel('x(n)')
plt.legend(['0.5','5','50'])
plt.show()

E = np.array([(1-np.exp(-2*a*N*1/fs))/(1-np.exp(-2*a*1/fs)) for a in A])

freq = fftfreq(N,d=1/fs)
for i,_x in enumerate(xn_seq):
    x = fft(_x)
    x = x**2 / E[i]
    plt.plot(freq,np.log(x))
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.legend(['0.5','5','50'])
plt.show()


```
